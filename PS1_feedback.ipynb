```python
#5.Path counting  5.2
def random_matrix(N,M):
    
    random_array=np.random.randint(0, 2, (N,M))
    matrix=random_array.reshape(N, M)
    
    matrix[0][0]=1
    matrix[N-1][M-1]=1
    
    return(matrix)


def Count_path(matrix):
    
    #确定矩阵第一个维度行的数量和第二个维度列的数量
    N=matrix.shape[0]
    M=matrix.shape[1]
    
    #定义函数用于运算每个格子的情况
    def recursive(i, j):
        #设置边界条件，如果当前单元格超出边界或0，则返回 0
        if i >= N or j >= M or matrix[i][j]==0:
            return 0
        #如果走到达右下角，则找到了一条路径
        if i==N-1 and j==M-1:
            return 1
        
        #向下和向右移动的路径总数
        down_paths=recursive(i + 1, j)
        right_paths=recursive(i, j + 1)
        
        #返回两种路径的总和
        return(down_paths+right_paths)

    #调用递归函数从最初位置开始计算路径
    return(recursive(0, 0))

N=10
M=8

total_paths_list=[]

for _ in range(1000):
    
    matrix=random_matrix(N,M)
    total_paths=Count_path(matrix)
    total_paths_list.append(total_paths)
    
    sum_total_paths_list=sum(total_paths_list)
    average=sum_total_paths_list/1000

print(total_paths_list)
print(sum_total_paths_list)
print(average)
```
